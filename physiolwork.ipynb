```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# サンプリングレート
sampling_rate = 1000

# ファイルパス
files = [
    'data1.csv',
    'data2.csv',
    'data3.csv',
    'data4.csv',
    'data5.csv'
]

for f, file in enumerate(files, start=1):
    # csvファイルの読み込み
    d = pd.read_csv(file).to_numpy()

    # Time window 解析区間（Eventから前後何秒？）
    if f == 1:
        tw_pre = 5 * sampling_rate
        tw_pst = 30 * sampling_rate
    elif f == 2:
        tw_pre = 10 * sampling_rate
        tw_pst = 180 * sampling_rate
    elif f == 3:
        tw_pre = 5 * sampling_rate
        tw_pst = 30 * sampling_rate
    elif f == 4:
        tw_pre = 5 * sampling_rate
        tw_pst = 15 * sampling_rate
    elif f == 5:
        tw_pre = 5 * sampling_rate
        tw_pst = 15 * sampling_rate

    # Event
    event_indices = np.where(d[:, 4] == 1)[0]
    if len(event_indices) == 0:
        print(f"No event found in file {file}")
        continue

    event_index = event_indices[0]
    start_index = max(event_index - tw_pre, 0)
    end_index = min(event_index + tw_pst, d.shape[0])

    BP = d[start_index:end_index, 1]
    MP = d[start_index:end_index, 2]
    HR = d[start_index:end_index, 3]

    # プロット
    fig, axs = plt.subplots(3, 1, figsize=(5, 10))
    yl_BP = [50, 200]
    yl_MP = [50, 200]
    yl_HR = [300, 500]

    axs[0].plot([tw_pre, tw_pre], yl_BP, 'k-')
    axs[0].plot(BP, 'r-')
    axs[0].set_ylim(yl_BP)
    axs[0].set_ylabel('Blood pressure (mmHg)')
    axs[0].set_xticks(np.arange(0, tw_pre + tw_pst, 5 * sampling_rate))
    axs[0].set_xticklabels(np.arange(-tw_pre / sampling_rate, tw_pst / sampling_rate, 5))

    axs[1].plot([tw_pre, tw_pre], yl_MP, 'k-')
    axs[1].plot(MP, 'k-')
    axs[1].set_ylim(yl_MP)
    axs[1].set_ylabel('Mean Blood pressure (mmHg)')
    axs[1].set_xticks(np.arange(0, tw_pre + tw_pst, 5 * sampling_rate))
    axs[1].set_xticklabels(np.arange(-tw_pre / sampling_rate, tw_pst / sampling_rate, 5))

    axs[2].plot([tw_pre, tw_pre], yl_HR, 'k-')
    axs[2].plot(HR, 'g-')
    axs[2].set_ylim(yl_HR)
    axs[2].set_ylabel('Heart rate (bpm)')
    axs[2].set_xticks(np.arange(0, tw_pre + tw_pst, 5 * sampling_rate))
    axs[2].set_xticklabels(np.arange(-tw_pre / sampling_rate, tw_pst / sampling_rate, 5))

    # 図の保存
    plt.tight_layout()
    plt.savefig(f'Fig_{f}.pdf')
    plt.show()
```
